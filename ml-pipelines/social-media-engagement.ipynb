```python
import pandas as pd

df = pd.read_csv('social_media_posts.csv')

df.head()
```

```python
import statsmodels.api as sm

features = [
    'platform',
    'post_type',
    'media_type',
    'post_hour',
    'day_of_week',
    'num_hashtags',
    'has_call_to_action',
    'sentiment_tone',
    'caption_length',
    'is_boosted',
    'features_resident_story'
]

df_model = df[features + ['engagement_rate']].dropna().copy()

X = df_model[features]
y = df_model['engagement_rate']

# Convert text → numbers
X = pd.get_dummies(X, drop_first=True)

# FORCE everything to numeric
X = X.astype(float)
y = y.astype(float)

# Add intercept
X = sm.add_constant(X)

# Train model
model = sm.OLS(y, X).fit()

print(model.summary())
```
